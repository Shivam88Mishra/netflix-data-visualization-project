```python
import pandas as pd
df= pd.read_csv('mymoviedb.csv', encoding='utf-8' , lineterminator='\n')
df.head()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Release_Date'] = df['Release_Date'].dt.year
df.head()
```

```python
df.drop(columns=['Original_Language', 'Poster_Url', 'Overview'], inplace=True, errors='ignore')
df.head()
```

```python
def catergorize_rating(df, col, labels):
    edges=[df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']]
    df[col]=pd.cut(df[col], edges, labels=labels, include_lowest=True , duplicates='drop')
    return df
labels=['flop', 'average', 'good', 'popular']
df=catergorize_rating(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
df['Genre']=df['Genre'].str.split(', ')
df=df.explode('Genre').reset_index(drop=True)
df.head()    
```
